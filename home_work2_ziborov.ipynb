## Импорт библиотек

```python
import pandas as pd
import numpy as np
```

## Создаем датафрейм из файла .csv

```python
df = pd.read_csv('titanic.csv')
df
```

## Пропущенные значения

```python
df.isnull().sum()
```

## Удаление всех строк, содержащих пропущенные значения в столбце 'embarked'

```python
df = df.dropna(subset=['embarked'])
```

Этот код заменит все пропущенные значения в столбце 'fare' на среднее значение этого столбца.
Аналогично можно использовать метод median для заполнения медианными значениями.

```python
mean_value_fare = round(df['fare'].mean(), 4)
print(mean_value_fare)
df['fare'].fillna(mean_value_fare, inplace=True)
```

## Удаление столбцов 'cabin', 'boat', 'body' из датафрейма

```python
df.drop(['cabin', 'boat', 'body'], axis=1)
```

## Замена отсутствующих значений в столбце 'home.dest' на текст 'not available'

```python
df.loc[df['home.dest'].isnull(), 'home.dest'] = 'not available'
df
```

## Мода:

```python
from statistics import mode
mode = mode(df['age'])

print(f'"модный" возраст среди пассажиров: {mode}')
```

## Среднее значение по столбцу 'age'

```python
mean_age = round(np.mean(df['age']), 2)

print(f'средний возраст пассажиров: {mean_age}')
```

## Распределение пассажиров по портам отправления

```python
df_embarked = df['embarked'].value_counts()
df_embarked
```

## Сохранение измененного датафрейма в новый файл.csv

```python
df.to_csv('titanic_edit.csv', index=False)
```
